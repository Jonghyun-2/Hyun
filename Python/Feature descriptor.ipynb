```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

### Haar 알고리즘<br>
---
1. Haar cascade classifier는 2001년 Paul viola 와 Michael jones의 논문 ‘rapid object detection using a boosted cascade of simple features’ 에서 제안된 효과적인 객체 검출 방법으로 알려져있다.<br>
2. 이미지 위에 검정색과 흰색영역으로 나뉜 마스크를 씌워 밝기 차를 이용하여 Feature를 추출해 내는 알고리즘이다.<br>
3. cascade 구조를 가지고 있는데 filter를 직렬구조로 놓고 맨 처음 단계에선 가장 object를 잘 찾는 filter를 찾아내고 그 다음 단계 에선 맨처음의 filter가 찾아낸 영역중 그보다 약한 filter를 사용하여 이러한 단계를 모두 거친 경우 object로 검출한다.




```python
haarFaceFilter = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

astro = data.astronaut()

plt.figure(figsize=(8,8))

plt.subplot(1,2,1)
plt.title("Orignal")
plt.imshow(astro)

faces = haarFaceFilter.detectMultiScale(astro)
for (x,y,w,h) in faces:
    cv2.rectangle(astro,(x,y),(x+w,y+h),(0,0,0),-1)

plt.subplot(1,2,2)
plt.title("Face detect")
plt.imshow(astro)

plt.show()
```
