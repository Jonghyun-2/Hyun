### Correlation analysis : 두 변수간 선형성을 가지고 있는지 분석하는 방법

### Correlation coefficient : 두 변수간의 연관된 정도

### Pearson correlation coefficient(Pearson's r) : Correlation coefficient를 구하기 위한 공식중 하나

```python
# Pandas 라이브러리 사용
import pandas as pd

lst = [[1,2,3,4,5,6,7],
      [10,20,30,40,50,60,70],
      [0,0,0,0,0,0,0],
      [-1,5,-10,24,34,67,52],
      [-1,-2,-3,-4,-5,-6,-7]]

df = pd.DataFrame(lst).T
cor = df.corr(method="pearson")

print(cor)

# lst[0] & lst[1] ==> perfect positive correlation
# lst[0] & lst[3] ==> strong positive correlation
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))

plt.subplot(231)
plt.title('lst[0]')
plt.plot(lst[0],'bo')

plt.subplot(232)
plt.title('lst[1]')
plt.plot(lst[1],'bo')

plt.subplot(233)
plt.title('lst[2]')
plt.plot(lst[2],'bo')

plt.subplot(234)
plt.title('lst[3]')
plt.plot(lst[3],'bo')

plt.subplot(235)
plt.title('lst[4]')
plt.plot(lst[4],'bo')
```

```python
import math
import numpy as np
```

```python
#pearson correlation coefficient
def pearson(lst0, lst1):
    lst0 = np.array(lst0)
    lst1 = np.array(lst1)
    lstSub = lst0 - lst1
    r = (np.sum(lstSub)) / (math.sqrt(np.sum(lstSub)**2))
    return r
```
