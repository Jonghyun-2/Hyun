```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
path = glob.glob(r'D:\Private\qq.jpg')
img = cv.imread(path[0])
biasRatio = 0.3

b = img[10,10,...][0]
g = img[10,10,...][1]
r = img[10,10,...][2]
bBias = b* biasRatio
gBias = g* biasRatio
rBias = r* biasRatio

chroma = cv.inRange(img,(b-bBias,g-gBias,r-rBias),(b+bBias,g+gBias,r+rBias))

dst0 = cv.bitwise_and(img,img,mask = chroma)
dst0 = cv.bitwise_not(img,dst0)
dst1 = cv.bitwise_or(img,img,mask = chroma)
plt.figure(figsize=(16,10))
plt.subplot(121)
plt.imshow(dst0)
plt.subplot(122)
plt.imshow(dst1)
# plt.subplot(121)
#img = img[:,:,::-1]
#plt.imshow(img)
```
